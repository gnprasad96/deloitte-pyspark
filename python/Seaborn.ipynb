# Seaborn

Seaborn is a Python data visualization library built on top of Matplotlib that provides a high-level interface for creating attractive and informative **statistical graphics**. 

It is widely used for visualizing data in a more visually appealing and informative manner compared to basic Matplotlib.

**Key Features of Seaborn:**

1. **Simplified Syntax**: Seaborn simplifies the creation of complex visualizations. For example, creating a heatmap, boxplot, or violin plot is much more concise with Seaborn.
1. **Built-in Themes**: Seaborn comes with several built-in themes to make your plots aesthetically pleasing with minimal effort. You can easily customize the look of your plots with functions like sns.set_theme().
1. **Statistical Plots**: Seaborn has built-in support for creating statistical plots like:
    - Box Plots: For comparing distributions between categories.
    - Violin Plots: For displaying the distribution of a dataset.
    - Pair Plots: For visualizing pairwise relationships in a dataset.
    - Correlation Heatmaps: For visualizing the correlation matrix of numerical variables.
1.**Works Well with Pandas**: Seaborn is designed to work well with Pandas DataFrames, making it easy to plot directly from data stored in these structures.
1. **Automatic Handling of Data**: It automatically calculates and displays statistical information like means, medians, confidence intervals, etc., for visualizations (e.g., in boxplots or regression plots).

#Histogram

A histogram is a classic visualization tool that represents the distribution of one or more variables by counting the number of observations that fall within disrete bins.

This function can normalize the statistic computed within each bin to estimate frequency, density or probability mass, and it can add a smooth curve obtained using a kernel density estimate

````python
penguins = sns.load_dataset("penguins")
sns.histplot(data=penguins, x="flipper_length_mm",kde=True)
penguins.head(5)
````

````python
sns.histplot(data=penguins, y="flipper_length_mm")
````

````python
sns.histplot(data=penguins, x="flipper_length_mm", binwidth=5)
````

````python
sns.histplot(data=penguins, x="flipper_length_mm", bins=30)
````

````python
sns.histplot(data=penguins, x="flipper_length_mm", kde=True)
````

````python
sns.histplot(data=penguins, x="flipper_length_mm", hue="species",multiple='dodge')#default multiple will be layer
````

### scatterplot
Draw a scatter plot with possibility of several semantic groupings.

The relationship between x and y can be shown for different subsets of the data using the hue, size, and style parameters. These parameters control what visual semantics are used to identify the different subsets. It is possible to show up to three dimensions independently by using all three semantic types, but this style of plot can be hard to interpret and is often ineffective. Using redundant semantics (i.e. both hue and style for the same variable) can be helpful for making graphics more accessible.

````python
tips = sns.load_dataset("tips")
tips.head(10)
````

````python
sns.scatterplot(data=tips, x="total_bill", y="tip")
````

````python
sns.scatterplot(data=tips, x="total_bill", y="tip", hue="time")
````

````python
sns.scatterplot(data=tips, x="total_bill", y="tip", hue="time", style="time")
````

### regplot

```regplot()``` is a function in Seaborn that creates a scatter plot with an optional regression line. It combines both the scatter plot and a linear regression line (or other types of regression) to help visualize the relationship between two continuous variables.

````python
import numpy as np; np.random.seed(8)
mean, cov = [4, 6], [(1.5, .7), (.7, 1)]
x, y = np.random.multivariate_normal(mean, cov, 80).T
ax = sns.regplot(x=x, y=y, color="g")
````

````python
ax = sns.regplot(x=x, y=y, ci=68, truncate=True)
````

````python
ax = sns.regplot(x="total_bill", y="tip", data=tips)
````

````python
ax = sns.regplot(x="size", y="total_bill", data=tips,
                 x_estimator=np.mean, logx=True)
````

### Jointplot

```jointplot()``` is a function in Seaborn used to visualize the relationship between two variables, while also showing their individual distributions. It provides a combination of scatter plots and histograms or kernel density estimates (KDE) to give insights into the correlation and distribution of the two variables

````python
penguins = sns.load_dataset("penguins")
print(penguins.columns)
print(penguins.info())
````

````python
sns.jointplot(data=penguins, x="bill_length_mm", y="bill_depth_mm")
````

````python
sns.jointplot(data=penguins, x="bill_length_mm", y="bill_depth_mm", hue="species")
````

````python
sns.jointplot(data=penguins, x="bill_length_mm", y="bill_depth_mm", hue="species", kind="kde")
````

````python
sns.jointplot(data=penguins, x="bill_length_mm", y="bill_depth_mm", kind="reg")
````

````python
sns.jointplot(data=penguins, x="bill_length_mm", y="bill_depth_mm", kind="hex")
````

### pairplot

```pairplot()``` is a function in Seaborn used to create a grid of scatter plots for each pair of variables in a dataset, along with histograms or KDEs along the diagonal to show the distribution of each individual variable. It is a very useful function for exploring relationships between multiple variables in a dataset.

````python
sns.pairplot(penguins)
````

````python
sns.pairplot(penguins, hue="species")
````

````python
sns.pairplot(
    penguins,
    x_vars=["bill_length_mm", "bill_depth_mm", "flipper_length_mm"],
    y_vars=["bill_length_mm", "bill_depth_mm"],
)
````

### countplot

```countplot() ```is a function in Seaborn used to display the counts of observations in each categorical bin using bars. It's a great way to visualize the distribution of categorical variables and show how many times each category appears in your dataset.

````python
sns.set_theme(style="darkgrid")
titanic = sns.load_dataset("titanic")
ax = sns.countplot(x="class", data=titanic)
````

````python
ax = sns.countplot(x="class", hue="who", data=titanic)
````
