# Matplotlib

**What is Matplotlib?**

Matplotlib is a popular Python library for data visualization. It provides functionalities to create a wide variety of static, interactive, and animated plots and charts. It is especially powerful when combined with other libraries like NumPy and pandas for data manipulation.

**Key Features**

- **Flexibility**: Allows customization of plots (e.g., colors, labels, sizes).
- **Wide Range of Plot Types**: Line plots, scatter plots, bar charts, histograms, etc.
- **Integration**: Works well with Jupyter notebooks and integrates seamlessly with other data science tools.

**Basic Steps to Plot a Graph Using Matplotlib**

1. *Install Matplotlib:* If you don't already have it installed, use:
```pip install matplotlib```
1. *Import Matplotlib:* Typically, you'll import the pyplot module as:
```import matplotlib.pyplot as plt```
1. *Prepare Data:* Define the data points you want to plot.
1. *Create a Plot:* Use plotting functions like plt.plot() for line plots, plt.scatter() for scatter plots, etc.
1. *Customize the Plot:* Add titles, labels, legends, etc.
1. *Show the Plot:* Use plt.show() to display the graph.

### Line Plot

````python
import matplotlib.pyplot as plt

# Data
x = [1, 2, 3, 4, 5]
y = [2, 4, 6, 8, 10]

# Create a plot
plt.plot(x, y, label='y = 2x', color='blue', linestyle='--', marker='o')

# Customize the plot
plt.title('Line Graph Example')
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.legend()  # Show the label as legend
plt.grid(True)  # Add grid

# Display the graph
plt.show()
````

**Demonstrating the use of ylim function**

```ylim``` in Matplotlib is used to set or get the limits of the y-axis in a plot. It controls the range of values displayed on the y-axis, allowing you to zoom in or out on specific sections of your data.

````python
fig = plt.figure(figsize=(10, 4), dpi=80)
ax1 = fig.add_subplot(1,2,1)
x=['Jan','Feb','March','April','May','June']
y2020=[55,33,77,99,22,55]
ax1.set_xlabel('Months')
ax1.set_ylabel('Temprature')
ax1.set_title('Month Vs Temp.')
ax1.plot(x,y2020,'gP--',label='Year 2020')
plt.legend()

# Demonstrating the use of ylim function

ax2 = fig.add_subplot(1,2,2)
y2020=[55,33,77,99,22,55]
ax2.set_ylim(20,80)
ax2.set_xlabel('Months')
ax2.set_ylabel('Temprature')
ax2.set_title('Month Vs Temp.')
ax2.plot(x,y2020,'gP--',label='Year 2020')
plt.legend()

plt.show()
````

### What is plt.figure?

In Matplotlib, plt.figure() is used to create a new figure (or canvas) for plotting. It allows you to:

- Create and manage multiple plots in the same script.
- Customize the figure's size, resolution, background color, etc.

**Key Parameters of plt.figure**
- figsize: Tuple (width, height) in inches, to specify the size of the figure.
- dpi: Dots per inch (resolution of the figure).
- facecolor: Background color of the figure.
- edgecolor: Edge color of the figure.
- num: Number or label for the figure.

**Example Simple plt.figure**

````python
# Create a figure with specific size
plt.figure(figsize=(8, 6), dpi=100)

# Plot data
x = [1, 2, 3, 4, 5]
y = [10, 20, 25, 30, 40]

plt.plot(x, y, label='Line 1')

# Add labels and title
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.title('Example with plt.figure')
plt.legend()

# Show the figure
plt.show()
````

**Example: Multiple Figures in One Script**

````python
# First figure
plt.figure(figsize=(6, 4), dpi=80)
x1 = [1, 2, 3]
y1 = [3, 6, 9]

plt.plot(x1, y1, 'r-', label='Red Line')
plt.title('Figure 1: Red Line')
plt.legend()

# Second figure
plt.figure(figsize=(6, 4), dpi=80)
x2 = [1, 2, 3]
y2 = [2, 4, 8]

plt.plot(x2, y2, 'b--', label='Blue Dashed Line')
plt.title('Figure 2: Blue Dashed Line')
plt.legend()

# Show both figures
plt.show()
````

**What is subplot in Matplotlib?**

In Matplotlib, *subplot* is a function that allows you to create multiple plots within a single figure (canvas). Each plot is called a subplot, and they can be arranged in a grid layout. This is useful for comparing different datasets or visualizations side by side.

Syntax of plt.subplot: ```plt.subplot(nrows, ncols, index)```

- nrows: The number of rows in the grid.
- ncols: The number of columns in the grid.
- index: The position of the current subplot (1-based index, left-to-right, top-to-bottom).

**Key Benefits of subplot**
- Efficient Use of Space: Compare multiple plots in a single figure.
- Organized Visualization: Arrange plots in rows and columns.
- Customization: Each subplot can have its own title, labels, and style.

**Example Basic Subplots:**

````python
# Create a figure with 2 rows and 1 column
plt.subplot(2, 1, 1)  # First subplot (row 1, column 1)
x1 = [1, 2, 3, 4]
y1 = [1, 4, 9, 16]

plt.plot(x1, y1, 'r-o')
plt.title('First Subplot')

plt.subplot(2, 1, 2)  # Second subplot (row 2, column 1)
x2 = [1, 2, 3, 4]
y2 = [4, 3, 2, 1]

plt.plot(x2, y2, 'b--x')
plt.title('Second Subplot')

plt.tight_layout()  # Adjust spacing between subplots
plt.show()
````

````python
plt.figure(figsize=(18, 6))

# Create a figure with 2 rows and 1 column
plt.subplot(1, 2, 1)  # First subplot (row 1, column 1)
x1 = [1, 2, 3, 4]
y1 = [1, 4, 9, 16]

plt.plot(x1, y1, 'r-o')
plt.title('First Subplot')

plt.subplot(1, 2, 2)  # Second subplot (row 2, column 1)
x2 = [1, 2, 3, 4]
y2 = [4, 3, 2, 1]

plt.plot(x2, y2, 'b--x')
plt.title('Second Subplot')

plt.tight_layout()  # Adjust spacing between subplots
plt.show()
````

**Example Subplot Grid Layout:**

````python
# Create a 2x2 grid of subplots
plt.subplot(2, 2, 1)  # Top-left
plt.plot([1, 2, 3], [1, 4, 9], 'r')
plt.title('Top Left')

plt.subplot(2, 2, 2)  # Top-right
plt.plot([1, 2, 3], [3, 2, 1], 'g')
plt.title('Top Right')

plt.subplot(2, 2, 3)  # Bottom-left
plt.plot([1, 2, 3], [2, 3, 2], 'b')
plt.title('Bottom Left')

plt.subplot(2, 2, 4)  # Bottom-right
plt.plot([1, 2, 3], [1, 2, 3], 'k')
plt.title('Bottom Right')

plt.tight_layout()  # Adjust spacing
plt.show()
````

**Example Common Title, X-axis, and Y-axis Labels:**

````python
# Create subplots
fig, axes = plt.subplots(2, 2, figsize=(10, 6))

# Add individual titles for subplots
axes[0, 0].plot([1, 2, 3], [4, 5, 6])
axes[0, 0].set_title("Top Left")

axes[0, 1].plot([1, 2, 3], [1, 2, 3])
axes[0, 1].set_title("Top Right")

axes[1, 0].plot([1, 2, 3], [3, 2, 1])
axes[1, 0].set_title("Bottom Left")

axes[1, 1].plot([1, 2, 3], [2, 3, 2])
axes[1, 1].set_title("Bottom Right")

# Add a common title
fig.suptitle("Common Title: Subplot Analysis", fontsize=16)

# Add common x-axis and y-axis labels
fig.text(0.5, 0.04, "Common X-axis Label", ha='center', va='center', fontsize=12)  # X-axis
fig.text(0.04, 0.5, "Common Y-axis Label", ha='center', va='center', rotation='vertical', fontsize=12)  # Y-axis

# Adjust spacing to prevent overlap
plt.tight_layout(rect=[0.05, 0.05, 1, 0.95])

# Show the figure
plt.show()
````

**Box Plots: To detect outliers and understand the spread of the data**

````python
df = diabetes

plt.figure(figsize=(12, 8))
plt.boxplot([df['Glucose'], df['BloodPressure'], df['SkinThickness'], df['Insulin'],
             df['BMI'], df['DiabetesPedigreeFunction'], df['Age']], 
            tick_labels=['Glucose', 'BloodPressure', 'SkinThickness', 'Insulin', 
                    'BMI', 'DiabetesPedigreeFunction', 'Age'])
plt.title('Boxplot of Each Column')
plt.show()
````

````python
# Select columns for pair plot (scatter plot matrix)
columns = ['Glucose', 'BMI', 'Age', 'BloodPressure']

plt.figure(figsize=(10, 8))

# Scatter plots of pairs of features
for i, col1 in enumerate(columns):
    for j, col2 in enumerate(columns):
        plt.subplot(len(columns), len(columns), i * len(columns) + j + 1)
        if i != j:
            plt.scatter(df[col1], df[col2], alpha=0.5)
            plt.xlabel(col1)
            plt.ylabel(col2)
        else:
            plt.hist(df[col1], bins=20)
            plt.xlabel(col1)
            plt.ylabel('Count')

plt.tight_layout()
plt.suptitle('Pairwise Relationships', y=1.02)
plt.show()
````
